Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill,Designations,Companies” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note- All of the above steps have to be done in code. No step is to be done manually.

```python
url='https://www.naukri.com/'
driver.get(url)

skills=driver.find_element_by_xpath("//input[@id='qsb-keyword-sugg']")
skills.send_keys('Data Analyst')

location=driver.find_element_by_xpath('//input[@id="qsb-location-sugg"]')
location.send_keys('Bangalore')

search=driver.find_element_by_xpath("//div[@class='search-btn']//button")
search.click()
```

```python
title_tags=driver.find_elements_by_xpath("//div/a[@class='title fw500 ellipsis']")
title_tags
job_titles=[]
for i in title_tags:
    job_titles.append(i.text)
job_titles

job_lctn=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
job_lctn
job_location=[]
for i in job_lctn[0:20]:
    job_location.append(i.text)
job_location

company=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company
company_name=[]
for i in company:
    company_name.append(i.text)
company_name

experience=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span")
experience
experience_req=[]
for i in experience:
    experience_req.append(i.text)
experience_req

match1=pd.DataFrame({})
match1['job-title']=job_titles
match1['job-location']=job_location
match1['company_name']=company_name
match1['experience_required']=experience_req
match1
```

Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Scientist” in “Skill,Designations,Companies” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note- 1. All of the above steps have to be done in code. No step is to be done manually.

```python
driver=webdriver.Firefox(executable_path=r"geckodriver.exe")
```

```python
url2="https://www.naukri.com/"
driver.get(url2)
time.sleep(3)

skills1=driver.find_element_by_id("qsb-keyword-sugg")
skills1.send_keys('Data Scientist')

location1=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
location1.send_keys('Bangalore')

search1=driver.find_element_by_xpath("//div[@class='search-btn']//button")
search1.click()
```

```python
allurls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    allurls.append(i.get_attribute('href'))
allurls
```

```python
jobs=[]
description=[]
location=[]
company1=[]
others1=[]

for i in allurls[:10]:
    driver.get(i)
    time.sleep(2)
    
    try:
        for j in driver.find_elements_by_xpath("//h1[@class='jd-header-title']"):
            jobs.append(j.text.replace("/n","new line"))
    except:
        jobs.append('-')
        
    try:
        for d in driver.find_elements_by_xpath("//section[@class='job-desc']"):
            description.append(d.text)
    except:
        description.append('-')

    try:
        for l in driver.find_elements_by_xpath("//span[@class='location ']"):
            location.append(l.text)
    except:
        location.append('-')
    
    try:
        cmpny=driver.find_elements_by_xpath("//a[@class='pad-rt-8']")
        for c in range(len(cmpny)):
            if c%2==0:
                company1.append(cmpny[c].text.replace('/n','new line'))
            else:
                others1.append(cmpny[c].text)
    except:
        company1.append('-')

match2=pd.DataFrame({})
match2['job-title']=jobs
match2['job-location']=location
match2['company_name']=company1
match2['full job-description']=description
match2
```
